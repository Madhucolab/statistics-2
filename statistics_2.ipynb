```python

                                                       # THEORY QUESTION
                                                      #  statistics-2

# 1.What is a random variable in probability theory?
ANS:- In probability a random variable is a function that assigns a numerical
value to each possible outcome of a random experiment, allowing us to quantify and analyze the
outcomes of chance processes.

Examples:
Discrete: The number of heads when flipping a coin three times (can only be 0, 1, 2, or 3).
Continuous: The height of a student in a class (can take any value within a range).

# 2.What are the types of random variables?
ANS:- Two Types are random variables:
1. Discrete Random Variable: Can only take on a finite or countably
infinite number of values (e.g., number of cars passing a point in an hour).
Examples include:
The number of heads when flipping a coin three times (can be 0, 1, 2, or 3).
The number of cars passing a certain point in an hour.
The number of defective items in a batch.
2. Continuous Random Variable: Can take on any value within a given range
(e.g., height, weight, temperature).

Examples include:
The height of a person.
The temperature of a room.
The weight of an object.
The time it takes to complete a task.

# 3.What is the difference between discrete and continuous distributions?

ANS:- Discrete Distributions:
1. Describe discrete random variables: These variables can only take
on a finite or countably infinite number of values.
2. Probability mass function (PMF): Used to describe the probability
of each possible value of the discrete variable.

Examples:
Bernoulli, binomial, Poisson, geometric distributions.
Continuous Distributions:
1. Describe continuous random variables: These variables can take
on any value within a given range.

2. Probability density function (PDF): Used to describe the
likelihood of the variable falling within a specific range.
Examples:
Normal, exponential, uniform distributions.

# 4.What are probability distribution functions (PDF)?

ANS:- In probability and statistics, a probability density function
(PDF) is a function that describes the probability distribution
of a continuous random variable. It represents the likelihood of the variable taking on a specific
value within a given range.

Example:
Consider the normal distribution (bell curve). The PDF of a normal distribution tells us the
probability of the variable taking on a specific value or falling within a certain interval. A higher PDF
value at a point indicates a higher likelihood of the variable being close to that value.

# Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

ANS:-Probability Density Function (PDF):
Focus:Describes the *relative likelihood* of a continuous
random variable taking on a specific value.
Output:The PDF provides a probability *density*, not a probability itself.
Interpretation:A higher PDF value at a particular point means the variable is more likely to
be found near that point, but it doesn't tell us the exact probability of it being exactly that value.
Example:Imagine the PDF of a student's height. A higher PDF value
at 175 cm would mean that heights around 175 cm are more common, but the probability of
a student being *exactly* 175 cm is technically zero (because there are infinitely many possible
heights).

Cumulative Distribution Function (CDF):
Focus:Gives the *probability* that a random variable is less
than or equal to a specific value.
Output:The CDF provides a probability value.
Interpretation: The CDF tells you the probability of the
variable being within a certain range (from negative infinity to a specific value).
Example:The CDF of a student's height at 175 cm would tell you the
probability that a randomly selected student's height is 175 cm or less.

# 6.What is a discrete uniform distribution?

ANS:- A discrete uniform distribution is a probability distribution
where each possible outcome has an equal probability of occurring. It's like rolling a fair die –
each face has a 1/6 chance of landing up.
Example:
- A fair six-sided die.Each number(1, 2, 3, 4, 5, 6) has a probability of 1/6.
- Picking a random card from a standard deck of cards. Each
card has a probability of 1/52.

# 7.What are the key properties of a Bernoulli distribution?

ANS:-The Key properties of a Bernoulli distribution:
Binary Outcomes:The random variable in a Bernoulli distribution
can only take on one of two possible values: 0 (representing failure) or 1 (representing success).
Single Trial:It models the probability of a single event or trial, not multiple trials.
Probability of Success (p):The distribution is characterized by a single parameter, 'p', which represents the
probability of success (i.e., the probability of the random variable taking the value 1).
Probability of Failure (1-p):The probability of failure (i.e.the random variable taking the value 0) is simply 1-p.
Independent Trials:Bernoulli trials are independent, meaning the
outcome of one trial does not influence the outcome of any other trial.
Foundation for Binomial Distribution:The Bernoulli distribution is the basis for the more general binomial
distribution, which models the probability of a certain number of successes in a fixed
number of independent Bernoulli trials.
Mean and Variance:The mean (expected value) of a Bernoulli
distribution is 'p', and the variance is 'p(1-p)

# 8.What is the binomial distribution, and how is it used in probability?

ANS:-The binomial distribution is a discrete probability distribution
that describes the probability of obtaining exactly k successes in a fixed number of independent
Bernoulli trials (e.g., coin flips, where each trial has only two possible outcomes).
Uses in Probability:
Modeling Events with Binary Outcomes: It's used in situations
where you want to find the probability of a specific number of successes (or failures) in a
fixed number of independent trials with a constant probability of success.
Quality Control: In manufacturing, it can be used to estimate
the probability of a certain number of defective products in a sample.
Medical Research: It can help determine the probability of a
certain number of patients responding positively to a new drug treatment.

Opinion Polls: It's employed to estimate the proportio of people with a certain opinion in a
population, based on a sample.
Genetics: It can help model the inheritance of certain traits,
where each offspring has a certain probability of inheriting a particular gene.

# 9.What is the Poisson distribution and where is it applied?

ANS:-The Poisson distribution is a discrete probability distribution
that models the probability of a certain number of events occurring within a fixed interval of time
or space, given the average rate of those events.
it applied:
It's commonly applied in fields like telecommunications, finance, and epidemiology to model
phenomena like customer arrivals, call center traffic, or disease outbreaks.

# Q10.What is a continuous uniform distribution?

ANS:-A continuous uniform distribution is a probability distribution
where every value within a specified range has an equal probability of occurring.
In simpler terms, it's like picking a random number from a continuous interval where each
number within that interval has the same chance of being chosen.
Example:
Imagine spinning a dial that can land on any value between 0 and 1. In a continuous uniform
distribution, each possible value between 0 and 1 has the same likelihood of being the outcome.

# 11.What are the characteristics of a normal distribution?

ANS:-characteristics of a normal distribution
Symmetry:The distribution is perfectly symmetrical around its
center, meaning the left side is a mirror image of the right side.
Unimodal:It has only one peak or mode, representing the most frequent data value.
Equal Mean, Median, and Mode: The mean, median, and mode all coincide at the center of the distribution.
Bell-Shaped Curve:The data points cluster around the mean, with
values further from the mean becoming less common.
Asymptotic Tails:The tails of the curve extend infinitely
towards the x-axis, never quite reaching it.
Standard Deviation:The standard deviation determines the spread
or width of the curve, indicating how much the data points deviate from the mean.
Central Limit Theorem: The central limit theorem states that the distribution of the sum (or
average) of a large number of independent, identically distributed variables will be approximately
normally distributed, regardless of the underlying distribution.

# 12.What is the standard normal distribution, and why is it important?

ANS:- The standard normal distribution is a specific type of normal
distribution with a mean of 0 and a standard deviation of 1. It's also known as the Z-distribution.
Importance:

1. Universal Comparison: It provides a standard scale for
comparing data from different normal distributions. By converting data to Z-scores
(number of standard deviations from the mean), you can easily compare values from datasets with
different means and standard deviations.

2. Probability Calculations: The standard normal distribution has
a well-defined probability density function (PDF) and cumulative distribution
function (CDF) with pre-calculated tables or software tools available. This enables us to easily
calculate the probability of a random variable falling within a specific range or exceeding a certain
threshold.

3. Hypothesis Testing: Many statistical tests rely on the
standard normal distribution for determining the probability of observing certain
sample results under a given hypothesis.

4. Central Limit Theorem Foundation: The central limit theorem,
which states that the sum or average of a large number of independent random
variables will be approximately normally distributed, is often used with the standard normal
distribution to make inferences about population parameters.

5. Data Transformations: In some cases, data that is not
normally distributed can be transformed to approximate a normal distribution
using techniques like log transformations or Box-Cox transformations. This can then allow the use of
statistical methods that rely on the normal distribution.

# Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

ANS:- The Central Limit Theorem (CLT) states that the sampling
distribution of the mean of a large number of independent, identically distributed random
variables will be approximately normally distributed, regardless of the shape of the original
population distribution.

Why it's critical:
Statistical Inference: The CLT allows us to make inferences about a population based on
sample data, even when we don't know the true population distribution.
Normal Distribution: It enables the use of normal distribution-based statistical methods (like
confidence intervals and hypothesis tests) for a wide range of data, as the distribution of sample
means will be approximately normal for large enough samples.
Generalizability: The CLT is a powerful tool because it holds true regardless of the shape of the
original population distribution, as long as the sample size is large enough.
Practical Applications: It's used in various fields, including finance, economics, and social
sciences, for estimating population parameters, making predictions, and testing hypotheses.

# Q14.How does the Central Limit Theorem relate to the normal distribution?

ANS:-The Central Limit Theorem (CLT) is closely related to the normal
distribution. Here's how:
1. The CLT states that the sampling distribution of the sample
mean from a population (regardless of the population's distribution) will approach a normal
distribution as the sample size increases.)

2. This means that even if the original data is not normally
distributed, the distribution of the means of many samples taken from that data will be
approximately normal, especially when the sample size is large.

3. The normal distribution becomes a crucial tool for making
inferences about population parameters based on sample statistics (like the sample mean)
because of the CLT.

4. The CLT allows us to apply many statistical methods that
assume normality, even when the original population is not normal, as long as the sample size
is large enough.

5. For example, we can use the normal distribution to construct
confidence intervals for the population mean, even if the population is not normally
distributed, because the CLT ensures that the sample mean distribution is approximately normal for
large sample sizes.

# Q15. What is the application of Z statistics in hypothesis testing?

ANS:- In hypothesis testing, the Z-statistic is a crucial tool used to
determine whether to reject or fail to reject the null hypothesis. It helps us assess whether the
observed sample mean is significantly different from a hypothesized population mean, considering
the standard deviation of the population.
Applications of Z-statistics:
It's widely used in hypothesis testing involving population means when the population standard
deviation is known.
It's used in quality control to determine whether a process is in control.
It's used in marketing research to test whether a new product is preferred over existing
products.

# Q16.How do you calculate a Z-score, and what does it represent?

ANS:- calculate a Z-score
Formula: z = (x - μ) / σ
Where:
z is the z-score
x is the data point you want to find the z-score for
μ (mu) is the mean of the dataset
σ (sigma) is the standard deviation of the dataset

Z-score represents:
A Z-score represents the number of standard deviations a data point is away from the mean of a
distribution.
A positive Z-score indicates that the data point is above the mean, while a negative Z-score
indicates that the data point is below the mean.
A Z-score of 0 means that the data point is equal to the mean.
Z-scores are useful for comparing data from different distributions, standardizing data, and
performing statistical tests.
Example:
If a test score of 85 has a z-score of 1.5, it means the score is 1.5 standard deviations above the
mean test score.

# 17. What are point estimates and interval estimates in statistics?

ANS:- Point Estimate:
1. A point estimate is a single value, calculated from a sample,
that is used to estimate the value of a population parameter.

2. For example, the sample mean (average) is a point estimate of
the population mean.

3. Point estimates are simple to calculate and communicate, but
they don't provide information about the accuracy or precision of the estimate.
Interval Estimate:
1. An interval estimate, such as a confidence interval, provides
a range of values within which the population parameter is likely to lie.
2. Interval estimates are more informative than point estimates
because they provide a measure of uncertainty or precision associated with the estimate.
3. For example, a 95% confidence interval means that there is a
95% probability that the true population parameter falls within the calculated interval.

# Q18.What is the significance of confidence intervals in statistical analysis?

ANS:-1. Quantifying Uncertainty:
They acknowledge that sample statistics are only estimates of the true population parameters.
There's always a degree of uncertainty or error associated with these estimates due to sampling variability.
Confidence intervals express this uncertainty by providing a range of plausible values for the population parameter.

2. Assessing Precision:
The width of the confidence interval reflects the precision of the estimate.
A narrower interval suggests a more precise estimate, while a wider interval indicates less precision (more uncertainty).

3. Making Inferences:
They enable us to make inferences about population parameters with a specified level of confidence.
For instance, a 95% confidence interval for the population mean suggests that if we were to
repeatedly sample from the population and calculate a confidence interval for each sample,
approximately 95% of these intervals would contain the true population mean.

4. Hypothesis Testing:
They are closely related to hypothesis testing.
If a confidence interval for a parameter does not include the null hypothesis value, it provides
evidence to reject the null hypothesis.

5. Decision Making:
Confidence intervals assist in decision-making.
For example, in a clinical trial, a confidence interval for the treatment effect can help
determine whether the treatment is effective, taking into account the uncertainty of the estimate.

6. Communication of Results:
They provide a clear and concise way to communicate the results of statistical analysis, along
with the level of uncertainty involved.

# Q19.What is the relationship between a Z-score and a confidence interval?

ANS:-Z-score:
A Z-score, also known as a standard score, indicates how many standard deviations an individual
data point is away from the mean of a dataset.
Relationship:
The Z-score, along with the standard deviation and sample size, is used to calculate the margin of
error, which is then added and subtracted from the sample mean to create the confidence interval.
Confidence Interval:
A confidence interval is a range of values within which a population parameter (like the mean) is
likely to fall, with a certain level of confidence.
Example:
For a 95% confidence interval, the Z-score is approximately 1.96, meaning that 95% of the data
falls within 1.96 standard deviations of the mean.

# Q20.How are Z-scores used to compare different distributions?

ANS:-Z-scores are used to compare different distributions by
standardizing the data.
This allows us to compare values from different distributions that may have different means and
standard deviations.
Here how Z-scores are used for comparing distributions:

1. Standardization:
Calculate the Z-score for each data point in each distribution.
This involves subtracting the mean of the distribution and dividing by the standard deviation.
The resulting Z-scores represent the number of standard deviations each data point is away from
the mean of its respective distribution.

2. Comparing Z-scores:
Now that all data points have been transformed to a standard scale (Z-scores), you can compare
values from different distributions more effectively.
For example, if you have two distributions, you can compare the Z-scores of two data points
from each distribution.
A higher Z-score means that the data point is farther away from the mean of its distribution,
relative to other data points in that distribution.
Example:
Suppose you have test scores from two different classes, each with their own mean and
standard deviation.
To compare a score of 80 from class A with a score of 75 from class B, you would:
1. Calculate the Z-score for the 80 in class A.
2. Calculate the Z-score for the 75 in class B.
3. Compare the Z-scores. The student with the higher Z-score
performed better relative to their respective class.

# Q21.What are the assumptions for applying the Central Limit Theorem?

ANS:-Assumptions for applying the Central Limit Theorem (CLT):
1. Independence: The random variables in the sample must be
independent of each other. This means that the outcome of one variable should not
affect the outcome of another.

2. Identically Distributed: The random variables should be drawn from the same population and
therefore have the same underlying distribution.

3. Sample Size: For the CLT to hold effectively, the sample size
should be sufficiently large. While there's no hard and fast rule, a sample size of at least
30 is often considered adequate for most distributions.

4. Finite Variance: The population from which the samples are drawn should have a finite
variance (or standard deviation). This means that the spread of the population data should not be
infinite.

5. If the population distribution is strongly skewed or has heavy
tails, you might need a larger sample size for the CLT to hold effectively.

# Q22.What is the concept of expected value in a probability distribution?

ANS:- Expected Value (E[X]) in a probability distribution represents
the average or mean outcome you would expect if you repeated an experiment or process many
times.
It's a weighted average of all possible outcomes, where each outcome is weighted by its
probability of occurrence.
Formula:
E[X] = Σ [x * P(x)]
Where:
x represents each possible outcome.
P(x) is the probability of that outcome occurring.
Example:
Let's consider a simple game where you roll a fair six-sided die.
The possible outcomes are 1, 2, 3, 4, 5, and 6, each with a probability of 1/6.
The expected value of this game is:
E[X] = (1 * 1/6) + (2 * 1/6) + (3 * 1/6) + (4 * 1/6) + (5 * 1/6) + (6 * 1/6) = 3.5

# Q23.How does a probability distribution relate to the expected outcome of a random variable?

ANS:- A probability distribution describes the likelihood of different
outcomes of a random variable.
The expected outcome of a random variable is the average value you would expect to get if you
repeated the experiment many times.
Here's how they relate:
1. Expected Value as a Weighted Average:
The expected value (E[X]) is calculated as a weighted average of all possible outcomes,
where each outcome is weighted by its probability of occurrence.

2. Probability Defines Weights:
The probability distribution provides the weights for each outcome. Higher probabilities
mean that the corresponding outcome contributes more to the expected value.

3. Reflecting Long-Term Average:
The expected value essentially represents the long-term average outcome you'd expect to
observe if you repeated the random experiment countless times.

Example:
Imagine flipping a biased coin where the probability of heads is 0.6 and tails is 0.4.
Let X be a random variable representing the outcome (1 for heads, 0 for tails).
The probability distribution is:
P(X=1) = 0.6
P(X=0) = 0.4
The expected value is:
E[X] = (1 * 0.6) + (0 * 0.4) = 0.6
This means that if you flipped the coin many times, you'd expect to get heads approximately 60%
of the time.
```

```python
# 2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low, high = 1, 6  # Range of values (e.g., a 6-sided die)

# Generate random samples from the uniform distribution
size = 1000  # Number of samples to generate
samples = np.random.randint(low, high+1, size=size)

# Plot the PMF
plt.figure(figsize=(8, 6))
values, counts = np.unique(samples, return_counts=True)

# PMF is the normalized frequency of each value
pmf = counts / size

plt.bar(values, pmf, width=0.8, align='center', color='skyblue', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xticks(values)  # Set the x-ticks to match the unique values
plt.show()
```

```python
# 3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(p, x):
    """
    Calculates the probability distribution function (PDF) for a Bernoulli distribution.

    Parameters:
    - p (float): Probability of success (1). Must be between 0 and 1.
    - x (int): The value to calculate the PDF for. Should be either 0 or 1.

    Returns:
    - float: Probability mass for the value x (either 0 or 1).
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be either 0 or 1")

# Example usage
p = 0.7  # Probability of success
x = 1  # We want the PDF for x = 1

pdf_value = bernoulli_pdf(p, x)
print(f"PDF at x = {x}: {pdf_value}")

```

```python
# 6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

# Parameters for discrete uniform distribution
a = 1  # Lower bound
b = 6  # Upper bound (inclusive)

# Discrete values for the uniform distribution
x = np.arange(a, b + 1)

# Probability for each value (uniform distribution, so each value has equal probability)
prob = np.ones_like(x) / (b - a + 1)

# Cumulative distribution function (CDF)
cdf = np.cumsum(prob)

# Plotting the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue', linewidth=2)
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('CDF')
plt.grid(True)
plt.xticks(x)  # Set x-ticks to the discrete values of x
plt.yticks(np.linspace(0, 1, 11))  # Set y-ticks from 0 to 1 with steps of 0.1
plt.show()
```

```python
# 9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Step 1: Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Step 2: Calculate the Z-scores
    z_scores = (data - mean) / std_dev

    # Step 3: Plot the Z-scores
    plt.figure(figsize=(8, 6))
    plt.plot(z_scores, marker='o', linestyle='-', color='b', label="Z-scores")
    plt.axhline(y=0, color='r', linestyle='--', label="Mean (Z=0)")
    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-Score")
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = [10, 12, 13, 15, 18, 19, 20, 21, 23, 25, 30]
z_scores = calculate_and_plot_zscores(data)
```

```python
# 10.mplement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(sample_size, num_samples, distribution="uniform", plot=True):
    # Step 1: Generate a non-normal distribution (Uniform distribution in this case)
    if distribution == "uniform":
        original_distribution = np.random.uniform(0, 10, 10000)  # Uniform distribution between 0 and 10
    else:
        raise ValueError("Currently only 'uniform' distribution is supported")

    # Step 2: Draw samples from the non-normal distribution and compute their means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(original_distribution, size=sample_size, replace=False)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Step 3: Plot the original distribution and the distribution of sample means
    if plot:
        plt.figure(figsize=(12, 6))

        # Plot original distribution
        plt.subplot(1, 2, 1)
        plt.hist(original_distribution, bins=50, color='skyblue', edgecolor='black')
        plt.title('Original Non-Normal Distribution (Uniform)')
        plt.xlabel('Value')
        plt.ylabel('Frequency')

        # Plot sample means distribution (CLT result)
        plt.subplot(1, 2, 2)
        plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
        plt.title(f'Distribution of Sample Means\n(Num Samples={num_samples}, Sample Size={sample_size})')
        plt.xlabel('Sample Mean')
        plt.ylabel('Frequency')

        plt.tight_layout()
        plt.show()

    # Return the distribution of sample means for analysis
    return sample_means
    # Example usage
sample_size = 30  # Number of elements in each sample
num_samples = 1000  # Number of samples drawn
sample_means = central_limit_theorem(sample_size, num_samples)

```

```python
# 15.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem_normal(population_size, sample_size, num_samples, mu=0, sigma=1, plot=True):
    # Step 1: Generate a normal distribution (population)
    population = np.random.normal(mu, sigma, population_size)

    # Step 2: Draw samples from the population and compute their means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Step 3: Plot the population distribution and the distribution of sample means
    if plot:
        plt.figure(figsize=(12, 6))

        # Plot population distribution (Normal distribution)
        plt.subplot(1, 2, 1)
        plt.hist(population, bins=50, color='skyblue', edgecolor='black')
        plt.title(f'Original Normal Distribution (μ={mu}, σ={sigma})')
        plt.xlabel('Value')
        plt.ylabel('Frequency')

        # Plot sample means distribution (CLT result)
        plt.subplot(1, 2, 2)
        plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
        plt.title(f'Distribution of Sample Means\n(Num Samples={num_samples}, Sample Size={sample_size})')
        plt.xlabel('Sample Mean')
        plt.ylabel('Frequency')

        plt.tight_layout()
        plt.show()

    # Return the distribution of sample means for analysis
    return sample_means

# Example usage
population_size = 10000  # Size of the population
sample_size = 30  # Number of elements in each sample
num_samples = 1000  # Number of samples drawn
sample_means = central_limit_theorem_normal(population_size, sample_size, num_samples)
```

```python
# 17.Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def generate_binomial_distribution(n, p, size):
    # Step 1: Generate random variables from a binomial distribution
    random_variables = np.random.binomial(n, p, size)

    # Step 2: Calculate the corresponding probabilities (PMF) for each value
    # The possible number of successes (k) ranges from 0 to n
    k_values = np.arange(0, n + 1)
    probabilities = binom.pmf(k_values, n, p)

    # Step 3: Plot the histogram of random variables and the binomial probability mass function
    plt.figure(figsize=(10, 6))

    # Histogram of generated random variables
    plt.hist(random_variables, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Generated Random Variables")

    # Plot the binomial distribution PMF
    plt.plot(k_values, probabilities, 'o', color='red', label="Binomial PMF", markersize=8)

    # Add labels and title
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")

    # Show grid and legend
    plt.grid(True)
    plt.legend()

    # Display the plot
    plt.show()

# Example usage
n = 10  # Number of trials
p = 0.5  # Probability of success in each trial
size = 1000  # Number of random variables to generate
generate_binomial_distribution(n, p, size)
```

```python
# 18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Function to calculate the Z-score for a given data point.

    :param data: List or array of data points (dataset).
    :param x: A single data point for which the Z-score needs to be calculated.
    :return: Z-score of the data point x.
    """
    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score using the formula: Z = (X - mean) / std_dev
    z_score = (x - mean) / std_dev
    return z_score

def plot_normal_distribution_and_zscore(data, z_score):

    # Generate values for plotting the standard normal distribution (mean=0, std=1)
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)  # Standard normal distribution (mean=0, std=1)

    # Plot the standard normal distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Plot the Z-score as a vertical line
    plt.axvline(x=z_score, color='red', linestyle='dashed', linewidth=2, label=f'Z-score: {z_score:.2f}')

    # Labels and title
    plt.title("Standard Normal Distribution and Z-Score Comparison")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")

    # Show legend and grid
    plt.legend()
    plt.grid(True)

    # Display the plot
    plt.show()

# Example usage:
data = [12, 15, 14, 10, 13, 17, 19, 20, 25, 16]  # Sample dataset
x = 18  # Data point for which we want to calculate the Z-score

# Calculate the Z-score for the given data point
z_score = calculate_z_score(data, x)

# Output the Z-score
print(f"The Z-score for the data point {x} is: {z_score:.2f}")

# Plot the comparison with the standard normal distribution
plot_normal_distribution_and_zscore(data, z_score)
```

```python
# 19.Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):

    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Step 2: Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Step 3: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Step 4: Make the decision
    reject_null = abs(z_statistic) > stats.norm.ppf(1 - alpha / 2)  # Critical Z-value for two-tailed test

    return z_statistic, p_value, reject_null

# Example usage
sample = [22, 24, 27, 29, 25, 23, 26, 21, 30, 28]  # Sample data
population_mean = 25  # Hypothesized population mean
population_std = 3   # Population standard deviation
alpha = 0.05         # Significance level

# Perform the Z-test
z_stat, p_val, reject_null = z_test(sample, population_mean, population_std, alpha)

# Output the results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_val:.4f}")
if reject_null:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# 20.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):

    # Step 1: Calculate sample mean, standard deviation, and sample size
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (Bessel's correction)
    sample_size = len(data)

    # Step 2: Calculate the Z-score for the given confidence level (for a normal distribution)
    # For large samples (n > 30), use Z-distribution (normal approximation)
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for the confidence level

    # Step 3: Calculate the margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    # Step 4: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
data = [22, 24, 27, 29, 25, 23, 26, 21, 30, 28]  # Sample dataset
confidence_level = 0.95  # 95% confidence level

# Calculate the confidence interval
lower, upper = confidence_interval(data, confidence_level)

# Output the result
print(f"The {int(confidence_level*100)}% confidence interval is: ({lower:.2f}, {upper:.2f})")
```

```python
# 23.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_poisson_cdf(lmbda, k_max):

    # Generate an array of k values from 0 to k_max
    k_values = np.arange(0, k_max+1)

    # Calculate the CDF for each k using the Poisson distribution
    cdf_values = stats.poisson.cdf(k_values, mu=lmbda)

    # Plot the CDF
    plt.figure(figsize=(8, 6))
    plt.step(k_values, cdf_values, where='post', label=f'Poisson CDF (λ={lmbda})', color='blue', linewidth=2)
    plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
    plt.xlabel('k (Number of events)')
    plt.ylabel('P(X ≤ k) (Cumulative Probability)')
    plt.grid(True)
    plt.legend()
    plt.show()

def poisson_cdf_interpretation(lmbda, k):

    # Calculate the Poisson CDF for the given k
    cdf_value = stats.poisson.cdf(k, mu=lmbda)

    # Interpretation
    print(f"The cumulative probability of observing {k} or fewer events is: {cdf_value:.4f}")

    # Interpretation text
    if cdf_value < 0.5:
        print(f"There is a less than 50% chance of observing {k} or fewer events.")
    else:
        print(f"There is a greater than 50% chance of observing {k} or fewer events.")

# Example usage:
lmbda = 3  # Average number of events (rate)
k_max = 10  # Maximum value of k for which to calculate the CDF

# Plot the Poisson CDF
plot_poisson_cdf(lmbda, k_max)

# Calculate and interpret the CDF for a specific k
k = 5
poisson_cdf_interpretation(lmbda, k)
```

```python
# 25.Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the standard deviation and compare
def compare_standard_deviations(data1, data2):

    # Step 1: Calculate the standard deviations
    std_dev_1 = np.std(data1, ddof=1)  # Sample standard deviation for data1
    std_dev_2 = np.std(data2, ddof=1)  # Sample standard deviation for data2

    print(f"Standard Deviation of Dataset 1: {std_dev_1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev_2:.2f}")

    # Step 2: Visualize the datasets and their spread
    plt.figure(figsize=(10, 6))

    # Create histograms for both datasets
    plt.hist(data1, bins=20, alpha=0.5, label='Dataset 1', color='blue', edgecolor='black')
    plt.hist(data2, bins=20, alpha=0.5, label='Dataset 2', color='red', edgecolor='black')

    # Add a vertical line for the mean of each dataset
    plt.axvline(np.mean(data1), color='blue', linestyle='dashed', linewidth=2, label=f'Mean of Dataset 1: {np.mean(data1):.2f}')
    plt.axvline(np.mean(data2), color='red', linestyle='dashed', linewidth=2, label=f'Mean of Dataset 2: {np.mean(data2):.2f}')

    # Step 3: Plot the standard deviation as a shaded area (optional)
    plt.fill_betweenx([0, 50], np.mean(data1) - std_dev_1, np.mean(data1) + std_dev_1, color='blue', alpha=0.1)
    plt.fill_betweenx([0, 50], np.mean(data2) - std_dev_2, np.mean(data2) + std_dev_2, color='red', alpha=0.1)

    # Step 4: Customize the plot
    plt.title('Comparison of Two Datasets')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)

    # Step 5: Show the plot
    plt.show()

# Example usage:
# Create two datasets with different spreads
np.random.seed(0)  # For reproducibility
data1 = np.random.normal(loc=50, scale=10, size=1000)  # Dataset 1 (mean=50, std=10)
data2 = np.random.normal(loc=50, scale=20, size=1000)  # Dataset 2 (mean=50, std=20)

# Compare the standard deviations and visualize the difference
compare_standard_deviations(data1, data2)
```

```python
# 26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np

# Function to calculate the range and interquartile range (IQR)
def calculate_range_and_iqr(data):

    # Step 1: Calculate the range (max - min)
    data_range = np.max(data) - np.min(data)

    # Step 2: Calculate the first and third quartiles (Q1 and Q3)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    # Step 3: Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    return data_range, IQR

# Example usage:
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Generate a normal distribution (mean=50, std=10)

# Calculate range and IQR
data_range, data_iqr = calculate_range_and_iqr(data)

# Print the results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {data_iqr:.2f}")
```

```python
# 27.Implement Z-score normalization on a dataset and visualize its transformation5
import numpy as np
import matplotlib.pyplot as plt

# Function to apply Z-score normalization
def z_score_normalization(data):

    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply the Z-score normalization formula
    normalized_data = (data - mean) / std_dev
    return normalized_data, mean, std_dev

# Example usage:
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Generate a normal distribution (mean=50, std=10)

# Step 1: Normalize the data using Z-score
normalized_data, mean, std_dev = z_score_normalization(data)

# Step 2: Visualize the original and normalized data
plt.figure(figsize=(12, 6))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Normalized Data (Z-score Normalization)')
plt.xlabel('Normalized Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()

# Print results for reference
print(f"Original Mean: {mean:.2f}")
print(f"Original Standard Deviation: {std_dev:.2f}")
print(f"Transformed (Normalized) Mean: {np.mean(normalized_data):.2f}")
print(f"Transformed (Normalized) Standard Deviation: {np.std(normalized_data):.2f}")
```

```python
# 28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# Function to calculate skewness and kurtosis of a dataset
def calculate_skewness_kurtosis(data):

    # Calculate skewness
    skew_value = skew(data)

    # Calculate kurtosis (excess kurtosis)
    kurt_value = kurtosis(data)

    return skew_value, kurt_value

# Example usage:
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Generate a normal distribution (mean=50, std=10)

# Step 1: Calculate skewness and kurtosis
skew_value, kurt_value = calculate_skewness_kurtosis(data)

# Step 2: Visualize the data to understand the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Step 3: Print results
print(f"Skewness of the dataset: {skew_value:.2f}")
print(f"Kurtosis of the dataset: {kurt_value:.2f}")
```
